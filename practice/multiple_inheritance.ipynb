```python
class Speaker:
    def say_hello(self):
        print("Hello there.")
class Driver:
    def drive(self):
        print("Driving...")
class SmartRobot(Speaker,Driver):
    def charge_battery(self):
        print("Battery charging..")
s=SmartRobot()
s.say_hello()
s.drive()
s.charge_battery()
```
