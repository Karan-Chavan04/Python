Dictionary
Dictionary is decalred using {}. It is a key value pair. Keys are unique and values are accessed through keys and assigned through keys. dict.values funtion is used to access list of values. dict.keys function is used to access the list of keys

```python
# Creating a dictionary to store user profile data
user_profile = {
    "username": "coder_karan",
    "language": "Python",
    "experience_level": "Intermediate",
    "completed_lessons": 4
}

# 1. How to access a value using its key
print("Current Language:", user_profile["language"])

# 2. How to update an existing value
user_profile["completed_lessons"] = 5

# 3. How to add a brand new key-value pair
user_profile["github_connected"] = True

print("\nUpdated Profile:", user_profile)
```

Set
Set is defined in {}. No duplicate values allowed, it is unordered , mutable.

```python
# Creating a set with duplicate items
languages_set = {"Python", "JavaScript", "C++", "Python", "C++"}

# Python automatically removes the duplicates!
print("Unique Languages:", languages_set)

# Adding a new item
languages_set.add("HTML")

# Checking if an item exists in a set (incredibly fast)
print("Is Python in the set?", "Python" in languages_set)
```
